```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
mot_a=pd.read_csv('Mot_A_rel_2.csv')
mot_b=pd.read_csv('Mot_B_rel_1.csv')
mot_a= mot_a[(mot_a['Temp']<180) &(mot_a['VM']<455)]
mot_b=mot_b[(mot_b['Temp']<180) &(mot_b['VM']<455)]
```

```python
sns.lineplot(data=mot_a,x=mot_a['r'],y=mot_a['T'])
```

```python
sns.lineplot(data=mot_a,x=mot_a['r']*mot_a['r'],y=mot_a['T'])
```

```python
corr = mot_a['r'].corr(mot_a['T'])
print(f"Pearson Correlation Coefficient: {corr}")
```

```python
corr1 = (mot_a['r']*mot_a['r']).corr(mot_a['T'])
print(f"Pearson Correlation Coefficient: {corr1}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(mot_a['r'], mot_a['T'], alpha=0.5)
plt.xlabel('r')
plt.ylabel('T')
plt.title('Scatter plot of T vs r')
plt.show()
```

```python
import statsmodels.api as sm

X = mot_a['r']
y = mot_a['T']
X = sm.add_constant(X)  # Add intercept
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
from sklearn.feature_selection import mutual_info_regression

mi = mutual_info_regression(mot_a[['r']], mot_a['T'])
print(f"Mutual Information Score: {mi[0]}")
```
